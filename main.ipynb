```python
source_path=os.path.join(SAMPLE_INPUTS,'sample.mp4')
frame_dir=os.path.join(SAMPLE_OUTPUTS,'fps')
all_frame_dir=os.path.join(SAMPLE_OUTPUTS,'all_frames')
output_vid=os.path.join(SAMPLE_OUTPUTS,'output.mp4')

os.makedirs(frame_dir,exist_ok=True)
os.makedirs(all_frame_dir,exist_ok=True)
```

```python
def frame_per_nthsec(n):
    for i,frame in enumerate(clip.iter_frames()): 
        fpns=int(fps/float(n))
        if (i % fpns)==0:
            image_save=os.path.join(all_frame_dir,f'{i}.jpg') 
            img=Image.fromarray(frame).save(image_save)

```

## Image to video


```python
inp_dir=os.listdir(all_frame_dir)
file_path=[] 
for fname in inp_dir:
    if fname.endswith('jpg'):
        paths=os.path.join(all_frame_dir,fname)
        file_path.append(paths)
        
```

```python
len(file_path)
```

```python
output_clip=ImageSequenceClip(file_path,fps=1)
```

```python
output_clip.write_videofile(output_vid)
```

## Here we sort the file paths to get ordered video

```python
dic={}

for root, dirs,files in os.walk(all_frame_dir):
    for fname in files:
        file_path=os.path.join(root,fname)
        key=float(fname.replace('.jpg',''))
        dic[key]=file_path

sorted_file_path=[]
for i in sorted(dic.keys()):
    path=dic[i]
    sorted_file_path.append(path)
    
```

## Ordered video saved as output.mp4
